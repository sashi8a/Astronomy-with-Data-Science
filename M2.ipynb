```python
dict = {'O':0,'B':0,'A':0,'F':0,'G':0,'K':0,'M':0}
total_MS = 0
```

```python
for e in df['SpType']:
    e = str(e)
    if (('V' in e) and ('I' not in e)):
        total_MS+=1
        
        if 'O' in e:
            dict['O']+=1
        if 'B' in e:
            dict['B']+=1
        if 'A' in e:
            dict['A']+=1
        if 'F' in e:
            dict['F']+=1
        if 'G' in e:
            dict['G']+=1
        if 'K' in e:
            dict['K']+=1
        if 'M' in e:
            dict['M']+=1
            

    
    
```

```python
plt.bar(dict.keys(), dict.values(), width = 0.6)
plt.xlabel('Spectral Class')
plt.ylabel('Raw Frequency')
plt.title('Distribution of Spectral Classes of Main Sequence Stars')
plt.show()
```

```python
Percentages = [(dict['O']/total_MS),(dict['B']/total_MS),(dict['A']/total_MS),(dict['F']/total_MS),(dict['G']/total_MS),(dict['K']/total_MS),(dict['M']/total_MS)]
plt.bar(dict.keys(), Percentages, width = 0.6)
plt.xlabel('Spectral Class')
plt.ylabel('Relative Frequency')
plt.title('Relative Frequencies of Spectral Classes of Main Sequence Stars')
plt.show()
```
