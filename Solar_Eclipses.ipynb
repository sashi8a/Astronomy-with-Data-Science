```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_solar = pd.read_csv('solar.csv')[:239]
df_solar
```

```python
sum_T = 0

for e in df_solar['Eclipse Type']:
    if e[0] =='T':
        sum_T+=1

sum_P = 0

for e in df_solar['Eclipse Type']:
    if e[0] =='P':
        sum_P+=1

sum_A = 0

for e in df_solar['Eclipse Type']:
    if e[0] =='A': 
        sum_A+=1     
        
sum_H = 0

for e in df_solar['Eclipse Type']:
    if e[0] =='H':
        sum_H+=1


        
totals = [sum_T,sum_P,sum_A,sum_H]


print(totals)
```

```python
types = ['Total','Partial', 'Annular','Hybrid']
c = ['red', 'yellow', 'green', 'blue']
plt.bar(types, totals, color = c, width = 0.6)
plt.xlabel('Type of Solar Eclipse')
plt.ylabel('Total Number of Eclipses')
plt.title('Distribution of Solar Eclipses')
plt.show()
```
