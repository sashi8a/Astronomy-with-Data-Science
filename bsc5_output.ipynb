```python
RA = (15*df['RAh']) + (15*df['RAm'])/60 + (15*df['RAs'])/3600

DEC = df['DEd'] + df['DEm']/60 + df['DEs']/3600

for i in range(len(DEC)):
    if df['DE-'][i]=='-':
        DEC[i]*=-1


Glon = df['GLON']
Glat = df['GLAT']
```

```python
plt.figure(figsize = (16, 8.4))
plt.subplot(projection = "aitoff")
plt.title("All-sky Plot in Equatorial Coordinates")
plt.plot(RA, DEC, 'o', markersize = 1, alpha = 0.85, color = 'b')
plt.grid(True)
plt.savefig('all_sky_eq.pdf')
plt.show()
```

```python
plt.figure(figsize = (16, 8.4))
plt.subplot(projection = "aitoff")
plt.title("All-sky Plot in Galactic Coordinates")
plt.plot(Glon, Glat, 'o', markersize = 1, alpha = 0.85, color = 'b')
plt.grid(True)
plt.savefig('all_sky_gal.pdf')
plt.show()
```
